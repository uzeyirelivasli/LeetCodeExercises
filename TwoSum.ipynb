Given an array of integers nums and an integer target, return indices of the two numbers such that they add up to target.

You may assume that each input would have exactly one solution, and you may not use the same element twice.

You can return the answer in any order.

 

Example 1:

Input: nums = [2,7,11,15], target = 9
Output: [0,1]
Explanation: Because nums[0] + nums[1] == 9, we return [0, 1].

```python
class Solution:
    def twoSum(self, List, target: int):
        # Dictionary to store numbers we've seen and their actual indices
        seen_numbers = {} 

        for current_index, number in enumerate(List):
            # Calculate the required pair to reach the target
            my_pair = target - number

            # If this pair was already seen, return its index and the current index
            if my_pair in seen_numbers:
                return [seen_numbers[my_pair], current_index]
            
            # Store the current number and its index in the dictionary
            seen_numbers[number] = current_index
```
